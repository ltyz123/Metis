```python
df = df.dropna()
from collections import Counter
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
import nltk
from nltk.corpus import stopwords
from nltk import sent_tokenize, word_tokenize
from wordcloud import WordCloud, STOPWORDS
import re

top_N = 100
#convert list of list into text
#a=''.join(str(r) for v in df_usa['title'] for r in v)
def clean_review(df):
    a = df['Review Text'].str.lower().str.cat(sep=' ')

    # removes punctuation,numbers and returns list of words
    b = re.sub('[^A-Za-z]+', ' ', a)

    #remove all the stopwords from the text
    stop_words = list(get_stop_words('en'))         
    nltk_words = list(stopwords.words('english'))   
    stop_words.extend(nltk_words)

    word_tokens = word_tokenize(b)
    filtered_sentence = [w for w in word_tokens if not w in stop_words]
    filtered_sentence = []
    for w in word_tokens:
        if w not in stop_words:
            filtered_sentence.append(w)

    # Remove characters which have length less than 2  
    without_single_chr = [word for word in filtered_sentence if len(word) > 2]

    # Remove numbers
    data = [word for word in without_single_chr if not word.isnumeric()]
    return data
```

```python
def get_neg(class_name):
    neg = []
    whole = []
    neg1 = df[(df['Rating']<4)&(df['Class Name']==class_name)&(df['Review Text'].str.contains("but"))&(df['Review Text'].str.contains('however')==False)]
    neg2 = df[(df['Rating']<4)&(df['Review Text'].str.contains("however"))&(df['Review Text'].str.contains('but')==False)]
    for review in neg1['Review Text']:
        whole.append(review.partition('but')) 
    for review in neg2['Review Text']:
        whole.append(review.partition('however'))
    for i in range(len(whole)):
        neg.append(whole[i][2])
    return neg    
```

```python
def get_pos(class_name):
    return df[(df['Rating']>3)&(df['Class Name']==class_name)]
```

```python
#0 for negative, 1 for positive
def get_right(sentiment, class_name):
    if sentiment == 0:
        return get_neg(class_name)
    if sentiment == 1:
        return get_pos(class_name)
```

```python
def get_topic(sentiment, class_name):
    n_components = 10

    dataset = get_right(sentiment, class_name)


    tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2,
                                    ngram_range=(1,5),
                                    max_features=1000,
                                    stop_words='english')
    tf = tf_vectorizer.fit_transform(dataset)
    lda = LatentDirichletAllocation(n_components=n_components, max_iter=5,
                                    learning_method='online',
                                    learning_offset=50.,
                                    random_state=0)
    lda.fit(tf)

    vocab = tf_vectorizer.get_feature_names()

    for topic in range(n_components):
        print(f"TOPIC {topic}")
        for j in np.argsort(lda.components_,1)[topic,:10]:
            print(vocab[j])
        print()
```

```python
def to_df(sentiment, class_name):
    if sentiment == 0:
        sentiment = 'Negative'
    if sentiment == 1:
        sentiment = 'Positive'
    title = class_name + '|' + sentiment
    reviews = get_right(sentiment, class_name)
    extracted_reviews = pd.DataFrame({title:reviews},index)
    return extracted_reviews
```
